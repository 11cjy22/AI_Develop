# 1、numpy

```python
import numpy as np
arr = np.array([1,2,3,4])
print(arr)
```

# 2、numpy的属性
## arr.shape: 数组的维度
## arr.ndim: 数组的轴
## arr.size: 数组的长度，即所有元素的个数
## arr.dtype: 每个元素的元素类型
## arr.itemsize: 每个元素的大小（字节数）
## type(arr): 数组的类型
## 两种写法: 1、对象名.属性  2、np.属性名(对象名)

```python
arr = np.arange(15).reshape((3,5 )) # arange(15)类似python的range(15),包左不包右，放在3个长度维度为5的一维数组中
print(arr)
```

# 3、创建ndarray对象

## 3.1、把Python数组封装成ndarray对象

```python
# int数组
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([1.2,3.4,5.6])
print(type(arr1))
print(type(arr2))
```

## 3.2、zeros()/ones()/empty()
### zeros: 创造一个全是0的数组
### ones: 创造一个全是1的数组
### empty: 创造一个内容随机的数组
### 默认创建的数组类型(dtype)都是float64

```python
zero1 = np.zeros((2,3))
print(f"zero1: {zero1}")
one1 = np.ones((2,3,4)) #3维数组：2个2维数组，每个2维数组有3个1维数组，每个1维数组有4个元素
print(f"one1: {one1}")
empty1 = np.empty((2,3))
print(f"empty1: {empty1}")
```

## 3.3、arange(起始值、结束值、步长、类型)函数
### arr = np.arange(1,5,2,dtype=np.int64): 生成1~5之间，步长为2的int64数据

## 3.4、matrix()生成二维数组
### arr = np.mat('1 2;3 4') : 生成 二维数组

```python
arr3 = np.arange(1,5,2,dtype=np.int64)
print("arr3: ",arr3)
print("类型：",arr3.dtype)
arr4 = np.mat('1 2; 3 4')
print("arr4:", arr4)
print("类型：",arr4.dtype)
```
